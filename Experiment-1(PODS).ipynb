```python
import pandas as pd
import numpy as np

df = pd.read_csv('Titanic-Dataset.csv')
```

```python
print(df.head(),"///")
print(df.info())
```

```python
df['Age_forward_fill'] = df['Age'].ffill()
df['Age_backward_fill'] = df['Age'].bfill()
df.head()
```

```python
df['Cabin'] = df['Cabin'].fillna("unknown")
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df['Sex_encoded'] = np.where(df['Sex'] == 'male', 1, 0)
df.head()
```

```python
fare_mean = df['Fare'].mean()
fare_std = df['Fare'].std()
df['Fare_scaled'] = (df['Fare'] - fare_mean) / fare_std
df.head()
```

```python
print(df[['Pclass', 'Sex_encoded', 'Age', 'SibSp']].head())
```

```python
corr = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
print(corr)
```
